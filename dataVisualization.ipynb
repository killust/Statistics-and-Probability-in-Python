```python
import pandas as pd
url = "datasets/nhanes/nhanes_2015_2016.csv"
da = pd.read_csv(url)

plt.figure(figsize = (20,5))

x = range(da['BMXWT'].size)
y = da['BMXWT']

plt.scatter(x, y, marker = 'x', s = 0.5)

plt.title("Weights of individuals")
plt.xlabel("ID of individuals")
plt.ylabel("Weight (kg)")

plt.show()
```
